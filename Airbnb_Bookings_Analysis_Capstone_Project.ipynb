```python
dir_path='/content/drive/MyDrive/Colab Notebooks/pandas/dat/'
df_air=pd.read_csv(dir_path+'Copy of Airbnb NYC 2019.csv')
df_air.head(3)
df_air.info()
```

```python
#create another dataframe having hosts,areas and their ratings,price,room type and avaialability
columns1=['host_id','host_name','neighbourhood','room_type','price','availability_365']
df_host_info=pd.DataFrame(df_air,columns=columns1)
df_host_info.head()
df_host_info.tail(20)
```

```python
#check the types of room
a=set(df_host_info.room_type)
print(f'Different types of rooms available \n{a}')

print('\n')

#check the different neighbourhood groups i.e state
n=set(df_air.neighbourhood_group)
print(f'Different neighbourhood groups i.e state \n{n}')

#There are different types of rooms available - 1. Private Room 2. Entire home/apt 3.shared rooms
# group by according to room and location
print('\n')
df_host_info.groupby(['room_type','neighbourhood']).count()
```

```python
#group according to different neighbourhood and price
#this gives about different prices on that location
df_host_info.groupby(['neighbourhood','price','room_type'])[['host_id']].count()

#Here in output we can see that for certain price offering how many hosts are present
```
